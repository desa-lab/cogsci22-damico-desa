```python
# Import synthetic data
data = pd.read_csv('fake_data.txt', delimiter='\t')

# Isolate C3 wave and time
time = data['time'].values
p3 = data['c3'].values

# Plot
plt.plot(time, p3);
plt.xlabel('Time (ms)');
plt.ylabel('Amplitude (AU)');
plt.title('Raw Idealized P3');
```

```python
# Filter data and plot compared to raw
fs = 200. # Hz (5ms between samples in time)
lowcut = 0.1
highcut = 10.
order = 2

# Generate second-order-section coeffs
# Note: you don't need to divide by nyquist if fs is passed
sos = signal.butter(order, [lowcut, highcut], analog=False, btype='band', output='sos', fs=fs)
p3_filt = signal.sosfiltfilt(sos, p3)

# Plot raw and filtered
plt.plot(time, p3, label='raw');
plt.plot(time, p3_filt, label='[0.1 10] filt');
plt.xlabel('Time (ms)');
plt.ylabel('Amplitude (AU)');
plt.title('Raw vs Filtered Idealized P3');
plt.legend();
```

```python
# Let's try it with a more complex waveform
p3_plus = data['c1+c2+c3+whitenoise']

p3_plus_filt = signal.sosfiltfilt(sos, p3_plus)

plt.plot(time, p3_plus, label='raw');
plt.plot(time, p3_plus_filt, label='[0.1 10] filt');
plt.xlabel('Time (ms)');
plt.ylabel('Amplitude (AU)');
plt.title('Raw vs Filtered Waveform');
plt.legend();
```
